```python
import numpy as np
import pandas as pd
from scipy.io import arff
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

import matplotlib.pyplot as plt # plotting
import random
random.seed(177)
rnd_state = np.random.seed(177)


df = arff.loadarff('C:/Users/jepey/Downloads/php9xWOpn.arff')
df = pd.DataFrame(df[0])
Y = df.loc[:, "Class"].astype("int")
X = df.iloc[:, 0:-1].astype("int")
df = (X, Y)

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.5, random_state=73
)

model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)


dataframe1= arff.loadarff('C:/Users/jepey/Downloads/php9xWOpn.arff') #SPF
dataframe1 = pd.DataFrame(dataframe1[0])
y1 = dataframe1.loc[:,"Class"].astype('int')
x1 = dataframe1.iloc[:,0:-1].astype('int')
dataframe1 = (x1,y1)

dataframe2= arff.loadarff('C:/Users/jepey/Downloads/dataset_59_ionosphere.arff') #Ionsphere
dataframe2 = pd.DataFrame(dataframe2[0])
y2 = dataframe2.loc[:,"class"].astype('str')
x2 = dataframe2.iloc[:,0:-1].astype('float')
dataframe2 = (x2,y2)

dataframe3= arff.loadarff('C:/Users/jepey/Downloads/php50jXam.arff') #Banknote
dataframe3 = pd.DataFrame(dataframe3[0])
y3 = dataframe3.loc[:,"Class"].astype('str')
x3 = dataframe3.iloc[:,0:-1].astype('float')
dataframe3 = (x3,y3)

datas = [dataframe1,dataframe2,dataframe3]

param1 = [1,2,3,4,5]
neighb1 = []
neighb2 = []
neighb3 = []
num = 0
for data in datas:
    x,y = data
    num = num + 1
    for i in param1:
        scores = []
        for rS in range(50):
            X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=rS+1)
            neigh = KNeighborsClassifier(n_neighbors=(i))
            neigh.fit(X_train, y_train)
            score = neigh.score(X_test,y_test)
            scores.append(score)
        if num == 1:
            neighb1.append(scores)
        elif num == 2:
            neighb2.append(scores)
        elif num == 3:
            neighb3.append(scores)






```
